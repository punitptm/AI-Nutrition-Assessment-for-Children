```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/child_health.csv")

df.head()
```

```python
# Convert height from cm to meters
df["height_m"] = df["height_cm"] / 100

# Calculate BMI
df["bmi"] = df["weight_kg"] / (df["height_m"] ** 2)

df[["age", "height_cm", "weight_kg", "bmi"]].head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(x="nutrition_status", data=df)
plt.title("Nutrition Status Distribution")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    x="age",
    y="bmi",
    hue="nutrition_status",
    data=df
)
plt.title("BMI vs Age")
plt.show()
```
